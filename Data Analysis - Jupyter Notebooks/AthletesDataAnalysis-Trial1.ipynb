### The other way to do this for 100% confidence is:

```python
def select_name(names_list):
    if len(names_list) == 1:
        return names_list
    else:
        name = None
        length_of_word = 0
        for each in names_list:
            if length_of_word < len(each):
                length_of_word = len(each)
                name = each
            else:
                continue
        return name

# Define the aggregation procedure outside of the groupby operation
aggregations = {
    'name': lambda x: select_name(x)
}
df_since_80_grouped = df_since_80.groupby(by=['db_resource_name']).agg(aggregations).reset_index()
```

```python
df_since_80_final = pd.merge(df_since_80, df_since_80_grouped, on='db_resource_name', how='inner')
```
