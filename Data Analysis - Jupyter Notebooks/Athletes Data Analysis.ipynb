# Athletes Data Analysis

```python
# User specification of file name
file_name = "athletes_11_02_2020_16_19_59.pkl"
```

```python
path = "../Project1_Athletes/Data/" + file_name
path
```

```python
#imports

# Basic imports
import pandas as pd
import numpy as np

#Visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Reading data from pickle file
df = pd.read_pickle(path)
df.head()
```

```python
# To check lang distributions in g.xml:lang and name.xml:lang
df['g.xml:lang'].value_counts() , df['name.xml:lang'].value_counts()
```

Since there are no other language apart from english, we can drop these columns.

```python
df.drop(columns=['g.xml:lang', 'name.xml:lang'], inplace= True)
df.head()
```

```python
# Renaming the columns
col_names_list = ["db_resource_name", "country", "dob", "gender", "sport", "height", "name"]
```

If this link is referenced, we can see their properties. For example, if this link http://dbpedia.org/page/Gerard_van_Velde is visitied, the properties of this dbpedia resource would come up and these properties that have been scraped can also be verified.

```python
df.columns = col_names_list
df.head()
```

```python
df.info(verbose=True, null_counts=True)
```

```python
# To remove "http://dbpedia.org/resource/" from every column

for col in ["db_resource_name", "country", "sport"]:
    df[col] = df[col].str.replace("http://dbpedia.org/resource/", "")

df.head()
```

Date of Birth has inconsistant date format and therefore there are duplicates from the source i.e. DBpedia

```python
# Make dob consistant
df[['dob_year', 'dob_month', 'dob_day']] = df.dob.str.split('-', expand=True)
df.head()
```

```python
# eyeballing the date of birth data

df.dob_year.value_counts(), df.dob_month.value_counts(), df.dob_day.value_counts()
```

We will be looking at dob range since the 1990 to current. Therefore, we store data only for dobs since 1990 in a new dataframe.

```python
# Type casting to integers as they are strings now
df['dob_year'] = df.dob_year.astype(int)
df['dob_month'] = df.dob_month.astype(int)
df['dob_day'] = df.dob_day.astype(int)

# Filetering dob year to begin from 1990
df_since_80 = df[df.dob_year > 1979]
len(df_since_80)
```

```python
df_since_80.dob_year.value_counts(), df_since_80.dob_month.value_counts(), df_since_80.dob_day.value_counts()
```

We see that the youngest player recorded on DBpedia was born in 2004. That makes sense because then current age would be ~16 years.

We see three incorrect data:
* 2985 as a birth year
* 0 as day of birth
* 0 as month of birth

Let's fix them!

```python
df_since_80 = df_since_80[~((df_since_80.dob_year == 2985) | (df_since_80.dob_day == 0) | (df_since_80.dob_month == 0))]
df_since_80.dob_year.value_counts(), df_since_80.dob_month.value_counts(), df_since_80.dob_day.value_counts()
```

Now let's check if few reknowned players are enlisted:

* Roger Federer: http://dbpedia.org/page/Roger_Federer
* Serena Williams: http://dbpedia.org/page/Serena_Williams
* Messi: http://dbpedia.org/page/Lionel_Messi
* Ronaldo: http://dbpedia.org/page/Cristiano_Ronaldo
* Saina Nehwal: http://dbpedia.org/page/Saina_Nehwal
* Chou Tien Chen: http://dbpedia.org/page/Chou_Tien-chen

```python
db_resource_name_list = ["Roger_Federer", "Serena_Williams", "Lionel_Messi", "Cristiano_Ronaldo", "Saina_Nehwal", "Chou_Tien-chen"]
for each_name in db_resource_name_list:
    print(df_since_80[(df_since_80.db_resource_name == each_name)])
```

We see that there are duplicates in terms of names (column name: name) and date of birth (column name: dob), we will:

* drop the dob column because we are retaining the data separately in year, month and day columns.
* grouping by db resource name, we will pick the maximum counted characters in name column and retain that one to remove duplicates

```python
df_since_80.drop(columns=['dob'], inplace=True)
df_since_80.head()
```
